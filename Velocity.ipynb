```python
#defind amount_1 function
def amount_1(df,name,col):
    '''
    Input name of dataframe, naming convention, and group by column name
    Will output dataframe with newly computed columns based on amount
    This function is intended to handle variable creation for Cardnum and Merchnum
    '''
    for d in day:
        df[name+"_Amount_mean_"+d]=df.groupby(col, as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].mean())
        df[name+"_Amount_max_"+d]=df.groupby(col, as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].max())
        df[name+"_Amount_median_"+d]=df.groupby(col, as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].median())
        df[name+"_Amount_total_"+d]=df.groupby(col, as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].sum())
        df[name+"_Amount_Act_mean_"+d]=df['Amount']/df[name+"_Amount_mean_"+d]
        df[name+"_Amount_Act_max_"+d]=df['Amount']/df[name+"_Amount_max_"+d]
        df[name+"_Amount_Act_median_"+d]=df['Amount']/df[name+"_Amount_median_"+d]
        df[name+"_Amount_Act_total_"+d]=df['Amount']/df[name+"_Amount_total_"+d]
    return df
```

```python
#defind amount_2 function
def amount_2(df,name1,name2,col1,col2):
    '''
    Input name of dataframe, naming convention, and group by column name
    Will output dataframe with newly computed columns based on amount
    This function is intended to handle variable creation for:
    Cardnum+ Merchnum,Cardnum+Merchzip, Cardnum+Merchstate
    '''
    for d in day:  
        df[name1+"_"+name2+"_Amount_mean_"+d]=df.groupby([col1,col2], as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].mean())
        df[name1+"_"+name2+"_Amount_max_"+d]=df.groupby([col1,col2], as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].max())
        df[name1+"_"+name2+"_Amount_median_"+d]=df.groupby([col1,col2], as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].median())
        df[name1+"_"+name2+"_Amount_total_"+d]=df.groupby([col1,col2], as_index=False, group_keys=False) \
                        .apply(lambda x: x.rolling(d,on="Date")['Amount'].sum())
        df[name1+"_"+name2+"_Amount_Act_mean_"+d]=df['Amount']/df[name1+"_"+name2+"_Amount_mean_"+d]
        df[name1+"_"+name2+"_Amount_Act_max_"+d]=df['Amount']/df[name1+"_"+name2+"_Amount_max_"+d]
        df[name1+"_"+name2+"_Amount_Act_median_"+d]=df['Amount']/df[name1+"_"+name2+"_Amount_median_"+d]
        df[name1+"_"+name2+"_Amount_Act_total_"+d]=df['Amount']/df[name1+"_"+name2+"_Amount_total_"+d]
    return df
```
